```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sb
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('sales_1.csv')
df.head()
```

```python
df.describe()
```

```python
df=df.drop(['SKU'], 1)
df.head()
```

```python

sb.countplot(x='Sales_pkg',data=df)
plt.xticks(fontsize=12)
plt.xticks(rotation=90)
plt.show();
```

```python
#plot of av
plt.figure(figsize=(20,10))
sb.countplot(x='Avg_price_pkg',data=df)
plt.xticks(fontsize=12)
plt.xticks(rotation=90)
plt.show();
```

```python
plt.figure(figsize=(20,10))
sb.countplot(x='Units_pkg',data=df)
plt.xticks(fontsize=12)
plt.xticks(rotation=90)
plt.show();
```

```python
#Plot of avg_price against group
plt.figure(figsize=(20,10))
sb.barplot(x='Group', y='Avg_price_pkg',data=df)
plt.xticks(fontsize=12)

plt.show();
```

```python
#Plot of Units_pkg against group
plt.figure(figsize=(20,10))
sb.barplot(x='Group', y='Units_pkg',data=df)
plt.xticks(fontsize=12)

plt.show();
```

```python
#Plot of sales_pkg against group
plt.figure(figsize=(20,10))
sb.barplot(x='Group', y='Sales_pkg',data=df)
plt.xticks(fontsize=12)

plt.show();
```
